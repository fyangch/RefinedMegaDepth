```python
import os
import numpy as np
import pandas as pd
```

```python
parent_dir = os.path.join("..", "..", "MegaDepthData", "Undistorted_SfM")
scenes = [dir for dir in os.listdir(parent_dir) if os.path.isdir(os.path.join(parent_dir, dir))]
print(f"{len(scenes)} scenes in total")
```

```python
# just a sanity check
valid = np.load(os.path.join(parent_dir, "valid.npy"))
print(f"numpy array contains {len(valid)} entries")

num_images = 0
for scene in scenes:
    image_dir = os.path.join(parent_dir, scene, "images")
    num_images += len(os.listdir(image_dir))
print(f"scene directories contain {num_images} images")
```

```python
valid_images = []
total_images = []
ratio = []

index = 0 # current index for the ´valid´ numpy array
for scene in scenes:
    image_dir = os.path.join(parent_dir, scene, "images")
    num_images = len(os.listdir(image_dir))
    
    num_valid_images = 0
    for i in range(index, index + num_images):
        if valid[i]:
            num_valid_images += 1
    
    valid_images.append(num_valid_images)
    total_images.append(num_images)
    ratio.append(num_valid_images / num_images)

    index += num_images


df = pd.DataFrame(
    index=scenes,
    data={
        "valid_images": valid_images,
        "total_images": total_images,
        "ratio": ratio,
    },
)
df = df.sort_values(by=["ratio"])
df
```
